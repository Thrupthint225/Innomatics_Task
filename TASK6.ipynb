```python
def smart_parking_system(capacity, logs):
    parked = 0  # current cars
    
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
    
    # decide status
    if parked > capacity:
        status = "Alert: Parking Capacity Exceeded!"
    elif parked == capacity:
        status = "Parking Full"
    else:
        status = "Available"
    
    print("Currently Parked Vehicles:", parked)
    print("Parking Status:", status)


# Example run
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking_system(capacity, logs)
```

```python
def delivery_time_estimator(distance_km, traffic, weather):
    # Base time: assume 5 minutes per km
    time = distance_km * 5  
    
    # Traffic delays
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5
    elif traffic == "Low":
        time += 0
    
    # Weather delays
    if weather == "Rainy":
        time += 5
    elif weather == "Stormy":
        time += 15
    elif weather == "Clear":
        time += 0
    
    print(f"Estimated Delivery Time: {time} minutes")
    return time


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

delivery_time_estimator(distance, traffic, weather)
```

```python
def seat_occupancy_analyzer():
    # Take inputs from user
    total_seats = int(input("Enter total seats in theatre: "))
    booked_count = int(input("Enter number of booked seats: "))
    
    # Step 1: Calculate occupancy percentage
    occupancy_percentage = (booked_count / total_seats) * 100
    
    # Step 2: Decide show status
    if booked_count == total_seats:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    elif occupancy_percentage >= 50:
        status = "Moderate Occupancy"
    else:
        status = "Low Occupancy - Consider Opening Additional Shows"
    
    # Step 3: Print results
    print(f"Occupancy: {int(occupancy_percentage)}%")
    print(f"Show Status: {status}")


# Run the function
seat_occupancy_analyzer()
```

```python
def server_load_classifier():
    # Step 1: Take input from user
    readings = input("Enter CPU readings (%) separated by spaces: ")
    
    # Convert input string into list of integers
    cpu_readings = list(map(int, readings.split()))
    
    # Step 2: Calculate average CPU usage
    avg_cpu = sum(cpu_readings) / len(cpu_readings)
    
    # Step 3: Classify server status
    if avg_cpu < 50:
        status = "Normal"
    elif 50 <= avg_cpu < 80:
        status = "Warning"
    else:  # avg_cpu >= 80
        status = "Critical"
    
    # Step 4: Print results
    print(f"Average CPU Load: {int(avg_cpu)}%")
    print(f"Server Status: {status}")


# Run the function
server_load_classifier()
```

```python
def resource_usage_monitor():
    # Step 1: Take input from user
    resources = {}
    n = int(input("Enter number of resources: "))
    
    for i in range(n):
        name = input(f"Enter resource {i+1} name: ")
        hours = int(input(f"Enter usage hours for {name}: "))
        resources[name] = hours
    
    # Step 2: Check for overuse (let’s assume >8 hours is overuse)
    overused = [res for res, hrs in resources.items() if hrs > 8]
    
    # Step 3: Energy alert if any resource is overused
    alert = "Yes" if overused else "No"
    
    # Step 4: Print results
    if overused:
        print("Overused Resources:", ", ".join(overused))
    else:
        print("Overused Resources: None")
    print("Energy Alert:", alert)


# Run the function
resource_usage_monitor()
```

```python
def event_registration_controller():
    # Step 1: Take inputs from user
    capacity = int(input("Enter event capacity: "))
    registrations = int(input("Enter number of registrations: "))
    
    # Step 2: Prevent overbooking
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    
    # Step 3: Print results
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Run the function
event_registration_controller()
```

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        # Check minimum amount
        if amount < 50:
            print("Recharge must be at least ₹50. Try again.")
            continue

        # Check if plan is valid
        if amount in valid_plans:
            print("Recharge successful! Plan activated.")
            break
        else:
            print("Invalid plan. Available plans are:", valid_plans)
            print("Please try again.")

# Call the function
validate_recharge()
```

```python
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"{item}: Reorder Alert")
        else:
            print(f"{item}: Stock OK")


# Example inventory dictionary
products = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 25,
    "Salt": 12
}

# Call the function
check_inventory(products)
```

```python
def process_result(marks):
    total = 0

    # Calculate total using loop
    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average} - Pass"
    else:
        return f"Average: {average} - Fail"


# Example marks list
student_marks = [60, 55, 40, 70]

# Call function
result = process_result(student_marks)
print(result)
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (distance * 12)

    # Add 25% extra during peak hours
    if peak.lower() == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    distance = float(input("Enter distance travelled (km): "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        print("Thank you for using the fare calculator.")
        break
```

```python
def check_attendance(attendance):
    present_days = 0

    # Count present days using loop
    for day in attendance:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    total_days = len(attendance)
    percentage = (present_days / total_days) * 100

    # Check eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example attendance list
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Call function
result = check_attendance(attendance_list)
print("Attendance Status:", result)
```

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Check each character using loop
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # Apply rules
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Example
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```

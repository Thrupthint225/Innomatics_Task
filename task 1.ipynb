```python
# User Login Check

# Predefined credentials
username = "admin"
password = "1234"

# Input from user
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Logic for authentication
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Pass / Fail Analyzer

# Marks list
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Logic to analyze results
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print final result
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
# Simple Data Cleaner

# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

# Clean and standardize names
cleaned_names = [name.strip().lower() for name in names]

# Print the cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
# Message Length Analyzer

# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")
    
    # Flag messages longer than 10 characters
    if length > 10:
        print("⚠️  This message is longer than 10 characters.\n")
    else:
        print("✅ This message is within limit.\n")
```

```python
# Error Message Detector

# System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Count the number of "ERROR" entries
error_count = logs.count("ERROR")

# Print the total error count
print("Total ERROR messages:", error_count)
```
